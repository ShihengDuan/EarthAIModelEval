```python
df_all = library.df_dict['rms_xyt']['ann']['global']

model_labels = list(df_all['model'])
model_names = model_labels
metric_names = ['pr', 'psl', 'ta-850', 'ta-200', 'va-850', 'va-200', 'zg-500']
data = df_all[metric_names].to_numpy()
model_highlights = None
print('data.shape:', data.shape)
print('len(model_names): ', len(model_labels))
```

```python
vertical_center = "median"
fig, ax = parallel_coordinate_plot(data, metric_names, model_names,
                                   title='Mean Climate rms_xyt',
                                   figsize=(21, 7),
                                   colormap='tab20',
                                   show_boxplot=False,
                                   show_violin=True,
                                   violin_colors=("lightgrey", "pink"),
                                   xtick_labelsize=10,
                                   logo_rect=[0.8, 0.8, 0.15, 0.15],
                                   comparing_models=model_highlights,
                                   vertical_center=vertical_center,
                                   vertical_center_line=True
                                  )

ax.set_xticklabels(metric_names, rotation=30, va='top', ha="right")
```

```python
df_all = combined.df_dict['rms_xyt']['ann']['global']
df_all = df_all[['model', 'pr', 'ta-850', 'ta-200', 'ua-850', 'ua-200', 'va-850', 'va-200', 'psl', 'zg-500', ]]
model_labels = list(df_all['model'])
model_names = model_labels
metric_names = ['pr', 'ta-850', 'ta-200', 'ua-850', 'ua-200', 'va-850', 'va-200', 'psl', 'zg-500', ]
data = df_all[metric_names].to_numpy()
model_highlights = None
print('data.shape:', data.shape)
print('len(model_names): ', len(model_labels))
```

```python
vertical_center = "median"
model_highlights = ['NeuralGCM-PCMDI', 'ACE2-PCMDI', 'NeuralGCM-evap']
fig, ax = parallel_coordinate_plot(data, metric_names, model_names,
                                   models_to_highlight=model_highlights,
                                   models_to_highlight_colors=['blue', 'red', 'green'],
                                   title='Mean Climate ANN rms_xyt',
                                   figsize=(18, 6),
                                   colormap='tab20',
                                   show_boxplot=False,
                                   show_violin=True,
                                   violin_colors=("lightgrey", "pink"),
                                   xtick_labelsize=12,
                                   logo_rect=[0.8, 0.8, 0.15, 0.15],
                                   comparing_models=model_highlights,
                                   vertical_center=vertical_center,
                                   vertical_center_line=True,
                                   identify_all_models=False,
                                   models_to_highlight_by_line=False, 
                                   # models_to_highlight_labels=None, 
                                   models_to_highlight_markers=['s', 'o', '^', '*'],
                                  )

ax.set_xticklabels(metric_names, rotation=30, va='top', ha="right")
plt.savefig('mean-climate-parallel.png', dpi=120, bbox_inches='tight')
```
